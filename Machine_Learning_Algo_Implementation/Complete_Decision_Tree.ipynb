#  Decision Tree Classifier Implementation With Post Prunning And Preprunning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
print(dataset.DESCR)
```

```python
import seaborn as sns
df=sns.load_dataset('iris')
```

```python
dataset.target
```

```python
#Independent and dependent features
x=df.iloc[:,:-1]
y=dataset.target
```

```python
### train test split
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
     x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier=DecisionTreeClassifier(criterion='entropy')
```

```python
classifier.fit(x_train,y_train)
```

```python
x_train.head()
```

```python
from sklearn import tree
plt.figure(figsize=(12,10))
tree.plot_tree(classifier,filled=True)
```

# Post Prunning

```python
classifier=DecisionTreeClassifier(criterion='entropy',max_depth=2)
classifier.fit(x_train,y_train)  # fitting the model
```

```python
from sklearn import tree
plt.figure(figsize=(12,10))
tree.plot_tree(classifier,filled=True)
```

```python
##prediction
y_pred=classifier.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,y_test)
```

```python
print(score)
print(classification_report(y_pred,y_test))
```

# DecisionTree Prepruning And Hyperparameter Tuning For Huge Data

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
parameter={
 'criterion':['gini','entropy','log_loss'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5],
  'max_features':['auto', 'sqrt', 'log2']

}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier=DecisionTreeClassifier()
clf=GridSearchCV(classifier,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
clf.fit(x_train,y_train)
```

```python
clf.best_params_
```

```python
y_pred=clf.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
score=accuracy_score(y_pred,y_test)
```

```python
print(score)
print(classification_report(y_pred,y_test))
```

# Decision Tree Regressor Implementation

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
##California House Pricing Dataset
from sklearn.datasets import fetch_california_housing
california_df=fetch_california_housing()
```

```python
california_df
```

```python
df=pd.DataFrame(california_df.data,columns=california_df.feature_names)
df['Target']=california_df.target
```

```python
df.shape
```

```python
## Taking Sample Data
df=df.sample(frac=0.25)
```

```python
df.shape
```

```python
#independent features
X=df.iloc[:,:-1]
#dependent features
y=df.iloc[:,-1]
```

```python
X.head()
```

```python
### train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
```

```python
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_pred,y_test)
```

```python
score
```

# Hyperparameter Tunning

```python
parameter={
 'criterion':['squared_error','friedman_mse','absolute_error','poisson'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5,6,7,8,10,11,12],
  'max_features':['auto', 'sqrt', 'log2']

}
regressor=DecisionTreeRegressor()
```

```python
from sklearn.model_selection import GridSearchCV
regressorcv=GridSearchCV(regressor,param_grid=parameter,cv=2,scoring='neg_mean_squared_error')
```

```python
regressorcv.fit(X_train,y_train)
```

```python
regressorcv.best_params_
```

```python
regressor=DecisionTreeRegressor(criterion='absolute_error',max_depth=7,
                                max_features='auto',splitter='best')
```

```python
regressor.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(12,10))
tree.plot_tree(regressor,filled=True)
plt.show()
```

```python
y_pred=regressorcv.predict(X_test)
```

```python
r2_score(y_pred,y_test)
```
